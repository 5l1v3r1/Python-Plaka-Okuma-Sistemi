#  Yıllara göre yoğunluk dağılımı

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt


yol='C:\\Users\\mertoglue\\AnacondaProjects\\CSV\\analiz\\gelen.csv'
        
veri = pd.read_csv(yol, encoding = "ISO-8859-1" ,sep=',')   # pandas ile dosya okunup veriler alınıyor ve okunan veriler, veri değişkenine atanıyor
filtre=veri['yil']

# filtre

yıllar = [2015,2016,2017,2018,2019]

plt.hist(filtre,yıllar ,width=0.5, label="Araç sayısı") # by default number of bins is set to 10

plt.ylabel('Araç sayısı')
plt.xlabel('Yıllar')
plt.title('Yıllar göre yoğunluk dağılımı')
plt.legend()
plt.xticks(yıllar)
plt.show()
```

# Yıllara göre araçların yüzdelik dilimde gösterimi

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt


yol='C:\\Users\\mertoglue\\AnacondaProjects\\CSV\\analiz\\gelen.csv'
        
veri = pd.read_csv(yol, encoding = "ISO-8859-1" ,sep=',')   # pandas ile dosya okunup veriler alınıyor ve okunan veriler, veri değişkenine atanıyor

onSekiz=veri[veri['yil']==2018].count().iloc[0]
onYedi=veri[veri['yil']==2017 ].count().iloc[0]
onAltı=veri[veri['yil']==2016 ].count().iloc[0]
onBes=veri[veri['yil']==2015 ].count().iloc[0]

liste=[]
liste.append(onSekiz)
liste.append(onYedi)
liste.append(onAltı)
liste.append(onBes)

activities = ['onSekiz','onYedi','onAltı','onBes']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

plt.pie(liste,
        labels=activities,
        colors=colors,
        startangle=90,        # 90 derece olarak başlatıyor
        shadow= True,
        explode=(0,0.1,0,0),  # dört yüzde değeri olduğu için parantez içindeki herbir değer bir yüzdeyi ifade eder
                              # değerin kendisi de dışa cıkıntı oranını veririr 
        autopct='%1.1f%%')

plt.show()
```
